# Tables d'intéret

Ici je ne garde que les tables qui m'interessent (stops, trips et stop_times). Je veux relier les parcours (route_id) aux arrets (stop_id). 


Je fais une jointure entre df_routes et df_stop_time par trip_id  pour lier les voyages et les horaires

ensuite le resultats est joint a la table df_stop sur le stop_id pour recuperer les coordonnées et les nom d'arrets 

Source https://gtfs.org/documentation/schedule/reference/

```python
# Je vais chercher lier les voyages et les horaires de passage
merged_step_1 = pd.merge(df_trips[['route_id',  'trip_id', 'direction_id', 'trip_headsign','wheelchair_accessible']],
                         df_stop_times[['trip_id', 'stop_id', 'stop_sequence', 'pickup_type',  'drop_off_type']], 
                         on='trip_id')

# Je rajoute les infos sur les arrets (lat, long, nom)
merged_step_2 = pd.merge(merged_step_1, df_stops, on='stop_id')

# Nettoyage finale et suppression des doublons
merged_arret_parcours = merged_step_2[[
    'route_id', 
    'direction_id', 
    'trip_headsign',
    'stop_id', 
    'stop_name', 
    'stop_lat', 
    'stop_lon', 
    'stop_sequence',
    'wheelchair_accessible', 
    'wheelchair_boarding',  
    'pickup_type',
    'drop_off_type'
]].drop_duplicates()

merged_arret_parcours['Parcours'] = merged_arret_parcours['route_id'].astype(str).str.split('-').str[-1] # route_id = 1-76 devient parcours = 76
```

# Parcours du RTC

```python
import geopandas as gpd

df_parcours = gpd.read_file('../output/PArcoursJson/Parcours.json') # fichier crée sur arcgis pro a partir d'un shapefile 
df_parcours['Parcours'] = df_parcours['Parcours'].astype(str)
df_parcours = df_parcours.drop_duplicates(subset=['Parcours', 'Type'])
df_parcours.head()
```

```python
len(df_parcours.Parcours.unique())
```

```python
df_final = pd.merge(merged_arret_parcours, df_parcours[['Parcours', 'Type']], on='Parcours', how='left')
df_final.head(3)
```

```python
df_final["Type"].unique()
```

```python
# Je ne garde que les parcours de type 'Régulier' et 'Métrobus' et aussi uniquement les arrêts où la descente est permise (drop_off_type == 0)
df_final = df_final[
    (df_final['Type'].isin(['Régulier', 'Métrobus'])) & 
    (df_final['drop_off_type'] == 0)
]

df_final = df_final.drop(columns=['wheelchair_accessible', 'wheelchair_boarding', 'pickup_type', 'drop_off_type'])
```

```python
gdf = gpd.GeoDataFrame(
    df_final, 
    geometry=gpd.points_from_xy(df_final.stop_lon, df_final.stop_lat),
    crs="EPSG:4326"  #  WGS84
)
```
